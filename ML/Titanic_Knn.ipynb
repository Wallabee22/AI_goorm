```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# test 데이터를 활용
df = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")
```

```python
#결측치 제거
df.loc[df['Age'].isnull()] = df["Age"].median()
df["Age"]
```

```python
sep = ['Pclass','Age','SibSp','Parch','Fare']

X = df[sep]
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
```

```python
scores = []
for i  in range(1,101):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    # 자체 테스트
    score = model.score(X_test,y_test)
    print('n_neighbors={}점수는 : {}'.format(score, i))
    scores.append(score)
```

```python
index = list(range(1,len(scores)+1))
score_df = pd.DataFrame(scores,index=index,columns=['result'])
#최적값
idx = score_df[score_df['result']== score_df.values.max()].index
print('''최적화 값은 : {}
순서는 : {}
'''.format(score_df.values.max(),idx.values))
```
