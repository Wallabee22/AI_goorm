```python
# 전체 방의 수를 위의 값을 기준으로 네 그룹으로 나눈다.
# A1 : 75~100  3151 ~
# A2 : 50~75   2127 ~ 3151 
# A3 : 25~50   1462 ~ 2127
# A4 : 0~25    ~1462
```

```python
train.loc[train['total_rooms']>=3152, 'total_rooms_level'] =1
train.loc[(train['total_rooms']> 2127) & (train['total_rooms'] <= 3151), 'total_rooms_level'] =2
train.loc[ (train['total_rooms']> 1462) & (train['total_rooms'] <= 2127), 'total_rooms_level'] =3
train.loc[ train['total_rooms']<= 1462, 'total_rooms_level'] =4
```

```python
### 새로운 컬럼 room_level 만들기
# 전체 방의 수를 위의 값을 기준으로 네 그룹으로 나눈다.
# A1 : 75~100  3151 ~
# A2 : 50~75   2127 ~ 3151 
# A3 : 25~50   1462 ~ 2127
# A4 : 0~25    ~1462

### 새로운 컬럼 room_level 만들기
bool_val = np.where( (train['total_rooms']> 3151) , True, False)
train.loc[bool_val, "room_level"] = 1
train['room_level'].head(15)
```
