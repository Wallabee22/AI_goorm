```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

x1 = np.array([0,1,2,3,])
y1 = np.array([13, 18 ,11 ,10])

x2 = np.array([0,1,2,3,])
y2 = np.array([32, 81 ,11 ,11])

x3 = np.array([0,1,2,3,])
y3 = np.array([320, 810 ,110 ,110])

x4 = np.array([0,1,2,3,])
y4 = np.array([30, 80 ,10 ,10])

ax1.plot(x1, y1, label = 'title1')
ax2.plot(x2, y2, label = 'title2')
ax3.plot(x3, y3, label = 'title3')
ax4.plot(x4, y4, label = 'title4')
plt.legend()

plt.show()
```

```python
x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.figure(figsize = (10, 12))
fig, axs = plt.subplots(2)
axs[0].plot(x,sin_y,'b--')
axs[1].plot(x,cos_y, 'r--')
plt.show()
```

```python
x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
sin2_y = 2 * np.sin(x)
cos2_y = 2 * np.cos(x)

#plt.figure(figsize = (10, 12))
fig, axs = plt.subplots(2,2, figsize = (10, 12))
axs[0,0].plot(x,sin_y,'b--')
axs[0,1].plot(x,cos_y, 'r--')
axs[1,0].plot(x,sin2_y,'b--')
axs[1,1].plot(x,cos2_y, 'r--')
plt.show()
```
