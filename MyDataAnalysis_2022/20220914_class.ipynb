```python
plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 2행 2열, 첫번째
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 2행 2열, 두번째
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 2행 2열, 세번째
plt.subplot(2, 2, 3)
plt.plot(x,y,'o')
plt.title("SALES")

#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 2행 2열, 네번째
plt.subplot(2, 2, 4)
plt.plot(x,y, 'o')
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set_style('whitegrid')

tips = sns.load_dataset("tips")
tips
```

```python
## 48개
## 막대그래프(countplot)
## Boxplot(boxplot)
## 히스토그램(histplot)
## 산점도
## heatmap
## 회귀직선(lmplot)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y='total_bill', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y='size', data=tips)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x = 'total_bill', y='tip', data=tips, size=7)
plt.show()
```
