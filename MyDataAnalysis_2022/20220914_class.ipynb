```python
plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 2행 2열, 첫번째
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 2행 2열, 두번째
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 2행 2열, 세번째
plt.subplot(2, 2, 3)
plt.plot(x,y,'o')
plt.title("SALES")

#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 2행 2열, 네번째
plt.subplot(2, 2, 4)
plt.plot(x,y, 'o')
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
