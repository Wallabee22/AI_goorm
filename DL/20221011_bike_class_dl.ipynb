```python
tr = pd.read_csv('bike_mod_tr.csv')
test = pd.read_csv('bike_mod_test.csv')
sub = pd.read_csv('sampleSubmission.csv')

tr.shape, test.shape, sub.shape
```

### 입력과 출력 지정

```python
input_col = ['hour', 'temp', 'dayofweek', 'workingday', 'season', 'weather']
labeled_col = ['count']

X = tr[input_col]
y = tr[labeled_col]
X_val = test[input_col]
```

```python
from sklearn.model_selection import train_test_split
```

```python
## 자체 학습, 자체 평가
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                        random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(30,input_dim=6, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss="mean_squared_error", optimizer='rmsprop')
model.fit(X_train, y_train, epochs=300, batch_size=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
pred = model.predict(X_val)
```

```python
sub['count'] = pred

sub.loc[sub['count']<0, 'count'] = 0
```
